```python
import re
from nltk import sent_tokenize
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktTrainer
```

```python
with open('future.txt', 'r', encoding='utf-8') as f:
      t = f.read()
```

```python
t[19937:21644] # выбранный кусочек текста
```

```python
splitted_text = [re.sub('[\n\xa0]', ' ', x) for x in sent_tokenize(t[19937:21644])]
len(splitted_text) # чистим от ненужных символов и смотрим количество предложений
```

```python
splitted_text # случаи с "…" в конце предложения не распознал
```

```python
part_t = re.sub("[\n\xa0]",' ', t[19937:21644])
part_t # подготавливаем кусочек текста для дальнейшего разбиения
```

```python
regex = r'(?<=[….!?])\s(?=[^а-я])'
reg = re.split(regex, part_t)
reg # используем регулярное выражение для разбиения текста
```

```python
print('Gold length - ', len(splitted_text))
print('Reg length - ', len(reg)) # с помощью регулярного выражения получилось правильно разбить текст, поэтому образцом разбиения теперь становится вариант reg с 52 предложениями
```

```python
gold = reg
```

```python
trainer = PunktTrainer()
trainer.INCLUDE_ALL_COLLOCS = True
trainer.train(t)
 
tokenizer = PunktSentenceTokenizer(trainer.get_params())
```

```python
tok = tokenizer.tokenize(part_t)
tok
```

```python
print('Gold length - ', len(gold))
print('Tok length- ', len(tok)) # встречаем те же 48 предложений...
```

```python
set(gold) ^ set(tok) # посмотрим, где встречаются ошибки
```

```python
tokenizer.PUNCTUATION # проблема с символом '…', пробуем исправить
```

```python
tokenizer.PUNCTUATION = tuple(list(tokenizer.PUNCTUATION) + ['…'])
tokenizer.PUNCTUATION # добавляем '…' в список символов
```

```python
tok_2 = tokenizer.tokenize(part_t) # пробуем еще раз
print(len(tok_2)) # тот же результат
set(gold) ^ set(tok_2) # проблема не разрешилась
```

```python
tp = 0
fp = 0
fn = 0

for sent in gold:
    if len(sent_tokenize(sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(gold)-1):
    sent1, sent2 = gold[i], gold[i+1]
    sent = ' '.join([sent1, sent2])
    if len(sent_tokenize(sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F1 - ', f1)
```

```python
tp = 0
fp = 0
fn = 0

for sent in tok:
    if len(sent_tokenize(sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(tok)-1):
    sent1, sent2 = tok[i], tok[i+1]
    sent = ' '.join([sent1, sent2])
    if len(sent_tokenize(sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F1 - ', f1)
```

```python
len(set(gold) & set(tok)) / len(set(gold) | set(tok)) # оценки по метрикам
```
